```python
m1=np.random.randint(0,1000,(5,2),dtype=int)
m2=np.random.randint(0,1000,(5,2),dtype=int)
print(m1)
print(m2)
```

```python
md=np.empty(shape=[0, 3],dtype=int)

for i in range(m1.shape[0]):
    for j in range(m2.shape[0]):
            md=np.append(md,[[i,j,int(np.linalg.norm(m1[i,:]-m2[j,:]))]],axis=0)  #purposefully converted to int

print(md)
print(euclidean_distances(m1,m2))
```

```python
print(md.shape)
smd=md[md[:,2].argsort()]
print(smd)
k=5
print()
print(smd[0:k])
```

```python
o=z=0
for i in range (k):
    for j in range (3):
        if smd[i][j]==0:
            z+=1
        elif smd[i][j]==1:
            o+=1
# print(z)
# print(o)
if(o>z):
    print(1)
else:
    print(0)
```

```python
path='diabetes.csv'
data=pandas.read_csv(path)

print(data.shape)
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
arr=data.values
X=arr[:,0:8]
Y=arr[:,8]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)

print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler().fit(X_train) 
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))#correct prediction=71+33=104
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
x=[]
y=[]
for i in range(1,25):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    print(str(i)+" "+str(accuracy_score(y_test, y_pred)))
    x.append(i)
    y.append(accuracy_score(y_test, y_pred))
plt.plot(x, y)
```

```python
# d=euclidean_distances(X_test,X_train)
y_pred=[]
for i in range(X_test.shape[0]):
    d=np.empty(shape=[0, 2])
    for j in range(X_train.shape[0]):
        d=np.append(d,[[y_train[j],np.linalg.norm(X_test[i,0:8]-X_train[j,0:8])]],axis=0)
    print(d)
    
    sd=d[d[:,1].argsort()]
    print(sd)
    z=o=0
    k=5
    for i in range(k):
        if sd[i][0]==0:
            z=z+1
        else:
            o=o+1
    if z>o:
        y_pred=np.append(y_pred,0)
    else:
        y_pred=np.append(y_pred,1)
    print()
print(y_pred)
```

```python
# print(y_test.shape)
# print(y_pred.shape)
print(classification_report(y_test, y_pred))
```
